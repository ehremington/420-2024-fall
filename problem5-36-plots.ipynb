# 5.36 Plots

Question 5.36 has some plots and I want to do them and also see how different the choices are between them.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(323, 373, 50)
def p1(temp):
    return(5.48*10**10*np.exp(-40660/8.31/temp))

def p2(temp):
    return(9.94*10**10*np.exp(-42920/8.31/temp))

fig, ax =  plt.subplots()

ax.plot(t, p1(t), '-', label='p1')
ax.plot(t, p2(t), '-', label='p2')
ax.hlines(101000, 323, 373, linestyles='dashed')
ax.hlines(1.234*10**4, 323, 373, linestyles='dashed')
ax.legend()
ax.set_xlabel('temp (K)')
ax.set_ylabel('pressure (Pa)')
```

The next part wants you to look at several locations and use the pressure there to estimate the boiling temperature. So...

Ogden = 1430m
Leadville = 3090
Mt. Whitney = 4420
Mt. Everest = 8850

and from problem 1.16 you know that pressure as a function of height is:
$$P(z) = P(0)*e^{\frac{mgz}{k_B T}} = P(0)*e^{\tfrac{Mgz}{R T}}$$

```python
def p(z, T):
    return(1.01*10**5*np.exp(-0.029*9.8*z/8.31/T))
```

```python
fig1, ax1 =  plt.subplots()

ax1.plot(t, p1(t), '-', label='p1')
ax1.plot(t, p2(t), '-', label='p2')
ax1.plot(t, p(0, t), label='sea level')
ax1.plot(t, p(1430, t), label='Odgen')
ax1.plot(t, p(3090, t), label='Leadville')
ax1.plot(t, p(4420, t), label='Whitney')
ax1.plot(t, p(8850, t), label='Everest')
# ax1.hlines(101000, 323, 373, linestyles='dashed')
# ax1.hlines(1.234*10**4, 323, 373, linestyles='dashed')
ax1.vlines(373, 20000, 101000, colors='C2')
ax1.vlines(369, 20000, 101000, colors='C3')
ax1.vlines(365, 20000, 101000, colors='C4')
ax1.vlines(361, 20000, 101000, colors='C5')
ax1.vlines(350, 20000, 101000, colors='C6')
ax1.legend(loc=3)
ax1.set_xlabel('temp (K)')
ax1.set_ylabel('pressure (Pa)')
```
